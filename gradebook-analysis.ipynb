```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, PowerTransformer
from scipy import stats
```

```python
def load_clean_data(path):
    gradebook = pd.read_excel(path)
    gradebook = gradebook[gradebook.iloc[:, 9] > 0]
    gradebook.columns = ['FirstName', 'Surname', 'Email', 
                        'Quiz1', 'Assignment1', 'Quiz1.1', 
                        'Project', 'CourseWorkTotal', 'FinalExams', 
                        'FinalGrade']
    return gradebook
exp_grdbk = load_clean_data("../input/gradebookanalysis/experimental_gradebook_final.xlsx")
ctrl_grdbk = load_clean_data("../input/gradebookanalysis/control_gradebook_final.xlsx")
```

```python
experimental_grades = exp_df['FinalGrade'].values
control_grades = ctrl_grdbk['FinalGrade'].values
grades = pd.DataFrame()
grades['Exp'] = experimental_grades
grades['Ctrl'] = control_grades
grades.head()
```

```python
sns.distplot(grades['Exp'], hist=False, rug=True, color='red')
plt.title("Distribution of experimental group grades")
```

```python
sns.distplot(grades['Ctrl'], hist=False, rug=True, color='blue')
plt.title("Distribution of control group grades")
```

```python
#Data should follow a normal distribution in t-test calculations. 
pt = PowerTransformer(method='yeo-johnson', standardize=True)
grades['Normalized_ctrl'] = pt.fit_transform(grades[['Ctrl']])
grades['Normalized_exp'] = pt.fit_transform(grades[['Exp']])
```

```python
sns.distplot(grades['Normalized_ctrl'], hist=False, rug=True, color='blue')
plt.title("Scaled Gradebook Distribution")
```

```python
alpha = 0.05
# T-TEST COMPUTATION on raw data
t_value_raw,p_value_raw=stats.ttest_ind(grades['Ctrl'], grades['Exp'])
# T-TEST COMPUTATUTION on normalized data
t_value_norm, p_value_norm = stats.ttest_ind(grades['Normalized_ctrl'], grades['Normalized_exp'])
```

```python
#test-statistic values
print("test statistic for raw data: " + str(t_value_raw))
print("test statistic for normalized data: " +str(t_value_norm))
```

```python
#p-values
print("P-value for raw data: " + str(p_value_raw))
print("P-value for normalized data: " + str(p_value_norm))
```

p_value > alpha hence fail to reject Ho i.e mean of control group = mean of exp group

```python
## Custom t-test for comparison
def get_t_test(grades_df, grp1, grp2):
    grades_df['Diff'] = grades_df[grp1] - grades_df[grp2]
    grades_df['Diff_squared'] = grades_df['Diff']*grades_df['Diff']
    diff_sum = grades_df['Diff'].sum()
    diff_squared_sum = grades_df['Diff_squared'].sum()
    t_test = (diff_sum/len(grades_df))/np.sqrt((diff_squared_sum-(diff_sum**2)/(len(grades_df))/(len(grades_df)-1)*len(grades_df)))
    return t_test
calculated_t_value_raw = get_t_test(grades, "Exp", "Ctrl")
calculated_t_value_raw
```

```python
calculated_t_value_norm = get_t_test(grades, "Normalized_exp", "Normalized_ctrl")
calculated_t_value_norm
```
